Data Cleaning


Handling Missing values

```python
auto_df.replace("?",np.nan,inplace=True)
```

```python
auto_df.isnull().sum()
```

```python
auto_df.dropna(subset="price",axis=0,inplace=True)
```

```python
auto_df["normalized-losses"].fillna(auto_df["normalized-losses"].astype(float).mean(),inplace=True)
```

```python
auto_df.isnull().sum()
```

```python
auto_df.dropna(subset=["num-of-doors","bore","stroke","horsepower","peak-rpm"],inplace=True)
```

```python
auto_df.isnull().sum()
```

```python
auto_df.shape
```

Changing the Data Types

```python
auto_df[["bore","normalized-losses","stroke","horsepower","peak-rpm","price"]]=auto_df[["bore","normalized-losses","stroke","horsepower","peak-rpm","price"]].astype(float)
```

```python
auto_df.dtypes
```

Removing Duplicates

```python
auto_df.drop_duplicates(inplace=True)
```

```python
auto_df.shape
```

Outliers


```python
auto_df=auto_df[(auto_df["price"]>=auto_df["price"].quantile(0.05)) & (auto_df["price"]<=auto_df["price"].quantile(0.95))]
```

```python
auto_df.shape
```

EDA



1. Distribution of Car Prizes.

```python
plt.hist(x=auto_df["price"],bins=10,edgecolor="black")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Distribution of Car Prices")
plt.show()
```

We intrepret from the histogram plot that maximum car price range in 5000-20000. 

2. Relation of Engine Size and Horsepower.

```python
sns.scatterplot(x="engine-size",y="horsepower",data=auto_df,hue="fuel-type")
plt.xlabel("Engine size")
plt.ylabel("Horsepower")
plt.title("Engine size vs Horsepower by fuel types")
plt.show()
```

From Scatter Plot we intrepret that with an increase in Engine Size There is increase in the horsepower for both Diesel as well as Gas Cars ie: they are directly proportional to each other.

3. Relation of Body Style with Price.

```python
sns.boxplot(x="body-style",y="price",data=auto_df)
plt.xlabel("Body Style")
plt.ylabel("Price")
plt.title("Body Style vs Price")
plt.show()
```

4.  How does the fuel economy (city-mpg and highway-mpg) vary between different body styles?

```python
mpg=["city-mpg","highway-mpg"]
mpg_mean=auto_df.groupby("body-style")[mpg].mean()
```

```python
mpg_mean.plot(kind="bar",rot=0)
plt.xlabel("Body Style")
plt.ylabel("Average MPG")
plt.title("Average City MPG ang Highway MPG versur Body Style")
plt.show()
```

Therefore we conclude that Hatchback Body style has highest average MPG both in city as well as on Highway.

5. How does fuel economy vary with different makers?

```python
mpg_mean1=auto_df.groupby("make")[mpg].mean()
```

```python
mpg_mean1.plot(kind="bar")
plt.xlabel("Make")
plt.ylabel("Average MPG")
plt.title("Average MPG vaersus the makers of car")
plt.show()
```

So we conclude that Chevrolet has highest fuel efficiency both in City as well as on Hoighway.

6. How does fuel efficiency vary with different fuel types?

```python
mpg_mean2=auto_df.groupby("fuel-type")[mpg].mean()
```

```python
mpg_mean2.plot(kind="bar",rot=0)
plt.xlabel("Fuel Type")
plt.ylabel("Average MPG")
plt.title("Average MPG versus Fuel Type")
plt.show()
```

So we conclude that diesel cars have higher average MPG both in city as well as on highways.

7. How does the length, width, height of the car vary with different make and body style?

```python
dim=["length","width","height"]
body_style_mean=auto_df.groupby("body-style")[dim].mean()
make_mean=auto_df.groupby("make")[dim].mean()
```

```python
body_style_mean.plot(kind="bar",rot=0)
plt.title("Body Style versur average dimensions")
plt.xlabel("Body Style")
plt.ylabel("Average dimensions")
```

So we conclude that Average length is highest in Wagon while the other two dimensions are nearly equal for all styles.

```python
make_mean.plot(kind="bar")
plt.title("make versus average Dimensions")
plt.xlabel("Make")
plt.ylabel("Average Dimensions")
```

So Jaguar makes the longest cars.

8. Fimd the everage Price of cars based on the number of Drive Wheels?

```python
sns.barplot(x=auto_df.groupby("drive-wheels")["price"].mean().index,y=auto_df.groupby("drive-wheels")["price"].mean().values)
plt.xlabel("Drive Wheels")
plt.ylabel("Average Price")
plt.title("Drive wheels versus Average price")
plt.show()
```

So we conclude that rwd cars have highest average price.
